```
#https://www.practicepython.org/exercise/2014/04/10/10-list-overlap-comprehensions.html

import random

a_length = random.randint(10,50) #wybiera randomowy intiger z zakresu <10,50>
b_length = random.randint(10,50)

i = 0
a = []
b = []
c = []

while i < a_length:
  a.append(random.randint(1,99))
  i+=1

i = 0

while i < b_length:
  b.append(random.randint(1,99))
  i+=1

a.sort()
b.sort()

if a_length < b_length:
  for element in a:
    if element in b: #sprawdza czy dany element jest na liście
      c.append(element)

if a_length > b_length:
  for element in b:
    if element in a: #sprawdza czy dany element jest na liście
      c.append(element)

if a_length == b_length:
  for element in a:
    if element in b: #sprawdza czy dany element jest na liście
      c.append(element)

print(a)
print(b)

c = list(dict.fromkeys(c)) #stworzenie z listy kluczy do słownika i przekonwertowanie ich na liste pozwala łatwo wyeliminować duplikaty z listy, gdyż klucze w słowniku nie mogą się powtarzać
print(c)
```
