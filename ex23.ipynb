```
#https://www.practicepython.org/exercise/2014/12/14/23-file-overlap.html

with open ('/content/happynumbers.txt', 'r') as file1:
  content1 = file1.read()

with open ('/content/primenumbers.txt', 'r') as file2:
  content2 = file2.read()

content1 = content1.split('\n')
content2 = content2.split('\n')

common = []

for number1 in content1:
  if number1 in content2:
    common.append(number1)

for number2 in content2:
  if number2 in content1:
    common.append(number2)

common = list(dict.fromkeys(common))

print('Those numbers are both - happy nad prime:')
print(common)
```
