```
#https://www.practicepython.org/exercise/2014/04/16/11-check-primality-functions.html

def isprimenumber(x):
  possible_divisors = []
  divisors = []
  y=2
  while y<x:
    possible_divisors.append(y)
    y+=1
  for number in possible_divisors:
    if x % number == 0:
      divisors.append(number)
  if x == 2:
    answer = '%d is not a prime number.' % (x)
  elif len(divisors)>0:
    answer = '%d is not a prime number.' % (x)
  else:
    answer = '%d is a prime number.' % (x)
  return answer

print('Welcome to IS-IT-PRIME-NUMBER game!')

while True:
  try:
    x = input('Please, give me a number: ')
    if x == 'x':
      print('Oh, okay - bye.')
      break
    x = int(x)
    print(isprimenumber(x)) #dlatego tak wygląda zapis bo return mówi co funkcja zwróci, czyli jakby jej wynik jest zmienną
  except ValueError:
    print('I said NUMBER.')
```
