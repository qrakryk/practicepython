```
#https://www.practicepython.org/exercise/2014/02/05/02-odd-or-even.html

while True:
  try:
    number = input('Give me a number: ')
    number = int(number)
    if number%4 == 0:
      print('%d is an even number and is multiple of 4.' % (number)) #%d = %digit
      break
    elif number%2 == 0:
      print('%d is an even number.' % (number))
      break
    else:
      print('%d is an odd number.' % (number))
      break
  except ValueError: 
    print('Number have to be in an intiger format, please try again.')
    
```

```
#https://www.practicepython.org/exercise/2014/02/05/02-odd-or-even.html
#extras2

while True:
  try:
    num = input('Give me a number: ')
    num = int(num)
    
    check = input('Give me a number to divide it by: ')
    check = int(check)

    if num%check == 0:
      print('Number %d can be divided by number %d completely.' % (num, check))
      solution = num/check
      print('Solution of such equation is: %d' % (solution))
      break
    else:
      print('Number %d can be divided by number %d completely.' % (num, check))
      solution = num/check
      print('Solution of such equation is: %f' % (solution))
      break
  except ValueError: 
    print('Number have to be in an intiger format, please try again.')
```
