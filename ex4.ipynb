```
#https://www.practicepython.org/exercise/2014/02/26/04-divisors.html
print('I will give you divisors of provided number.')
while True:
  try:
    number = input('Give me a number: ')
    number = int(number)
    i = 1
    divisors = []
    while i<=number:
      if number%i == 0:
        divisors.append(i)
      i+=1  
    break    
  except ValueError:
    print('Number have to be an intiger type.')

print('Divisors of %d number are:' % (number))
print(divisors)
```
