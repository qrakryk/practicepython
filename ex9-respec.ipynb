```python
import random #human choice

def checkResult(choiceHuman):
  if choiceHuman == choicePC:
    print("Ohhh, you made it!")
    return 0
  elif choiceHuman > choicePC: 
    print("Your number is too big!")
    return 1
  else:
    print("Your number is too small!")
    return 1

def pickingPC(x,y):
  print("Welocme in guessing game!")
  print("Just let me pick my number...")
  choicePC = random.randint(x,y) #PC choice
  print("*bzzz, burp, beep, beep, boop*")
  print("Okay, I have one!")
  print("Guessing started!")
  return choicePC
```

```python
choicePC = pickingPC(1,999)

tries = 1 #count tries
while True:
  choiceHuman= input('Give me your guess: ')
  choiceHuman = int(choiceHuman) #human choice
  tries+=1
  if checkResult(choiceHuman) == 0:
    break

print("It took %d tries." % tries)

```
