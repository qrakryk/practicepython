```
#https://www.practicepython.org/exercise/2015/11/26/27-tic-tac-toe-draw.html

import random
def whowon(game):
  if (game[0][0] == 'X') and (game[0][1] == 'X') and (game[0][2] == 'X'): #lista list
    print('Player 1 won the game.')
  elif (game[0][0] == 'X') and (game[1][0] == 'X') and (game[2][0] == 'X'):
    print('Player 1 won the game.')
  elif (game[0][0] == 'X') and (game[1][1] == 'X') and (game[2][2] == 'X'):
    print('Player 1 won the game.')
  elif (game[2][0] == 'X') and (game[1][1] == 'X') and (game[0][2] == 'X'):
    print('Player 1 won the game.')
  elif (game[0][1] == 'X') and (game[1][1] == 'X') and (game[2][1] == 'X'):
    print('Player 1 won the game.')
  elif (game[0][2] == 'X') and (game[1][2] == 'X') and (game[2][2] == 'X'):
    print('Player 1 won the game.')
  elif (game[0][1] == 'O') and (game[1][1] == 'O') and (game[2][1] == 'O'):
    print('Player 2 won the game.')
  elif (game[0][2] == 'O') and (game[1][2] == 'O') and (game[2][2] == 'O'):
    print('Player 2 won the game.')
  elif (game[0][0] == 'O') and (game[0][1] == 'O') and (game[0][2] == 'O'):
    print('Player 2 won the game.')
  elif (game[0][0] == 'O') and (game[1][0] == 'O') and (game[2][0] == 'O'):
    print('Player 2 won the game.')
  elif (game[0][0] == 'O') and (game[1][1] == 'O') and (game[2][2] == 'O'):
    print('Player 2 won the game.')
  elif (game[2][0] == 'O') and (game[1][1] == 'O') and (game[0][2] == 'O'):
    print('Player 2 won the game.')
  else:
    print('Noone won the game.')

game = [[0, 0, 0],
	[0, 0, 0],
	[0, 0, 0]]


print('Lets play Tic Tac Toe game.')
print('I will ask you about row and column where you want to place X.')

#Player1
i = 0
moves = 0
while i < 3:
  while True:
    row1 = input('Row: ')
    column1 = input('Column: ')
    row1 = int(row1)
    column1 = int(column1)
    if game[row1-1][column1-1] == 0:
      moves+=1
      break
    else:
      print('That place is already used.')
  game[row1-1][column1-1] = 'X'
  print('Gameboard %d: ' % moves)
  for row in game:
    print(row)

  #Player2
  row2 = random.randint(0,2)
  column2 = random.randint(0,2)

  while True:
    if game[row2][column2] != 0:
      row2 = random.randint(1,2)
      column2 = random.randint(1,2)
    else:
      moves+=1
      break

  game[row2][column2] = 'O'
  print('Gameboard %d: ' % moves)
  for row in game:
    print(row)
  i+=1

whowon(game)
```
