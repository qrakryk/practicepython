```
#https://www.practicepython.org/exercise/2017/01/10/32-hangman.html

import random

with open('/content/sample_data/sowpods.txt','r') as filename:
  filecontent = filename.read()
  filecontent = filecontent.split('\n')
  word = random.choice(filecontent)

print(word)

word = list(word)
word_len = len(word)
gameboard = []
guessed = []

for letter in word:
  gameboard.append('_ ')
gameboard_display = ''.join(gameboard)

tryp = 0
tryc = 0
tryi = 0

print('Welocme to hangman game!')
print('I choosed a word, please guess it letter by letter.')
print('You can guess incorrectly only 6 times.')

while True:
  gameboard_display = ''.join(gameboard)
  print(gameboard_display)
  print('\n')

  guess = input('Letter: ')
  guess = guess.upper()

  if guess in guessed:
    print('You already tried that letter.')
  elif guess in word:
    print('Correct guess: %s' % guess)
    tryp+=1
    tryc+=1
    i = 0
    for letter in word:
      if guess == letter:
        gameboard[i] = guess
      i+=1
  else:
    print('Incorrect guess: %s' % guess)
    tryp+=1
    tryi+=1
    if tryi == 1:
      print("/")
    elif tryi == 2:
      print(r"/ \ ")
    elif tryi == 3:
      print(' | ')
      print(r"/ \ ")
    elif tryi == 4:
      print(' | ')
      print(' | ')
      print(r"/ \ ")
    elif tryi == 5:
      print('  _ ')
      print(' | ')
      print(' | ')
      print(r"/ \ ")
    elif tryi == 6:
      print('  _ ')
      print(' | O')
      print(' | ')
      print(r"/ \ ")

    print('You have %d incorrect guesses left.' % (6-tryi))
    
  guessed.append(guess.upper())

  if tryi == 6:
    print('You lost the game.')
    break

  if gameboard == word:
    print('\n')
    print('Congratulations, you won and it took %d attempts.' % tryp)
    print('Correct guesses: %d' % tryc)
    print('Incorrect guesses: %d' % tryi)
    break
```
