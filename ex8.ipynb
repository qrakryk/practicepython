```
#http://www.practicepython.org/exercise/2014/03/26/08-rock-paper-scissors.html

import random

print('We are going to play a game! Remember, type "stop" if you would like to end the game.')

moves = ['rock', 'paper', 'scisors']

while True:
  player2 = random.choice(moves) #wybiera losowy element z listy i przypisuje do zmiennej "player2"
  player1 = input('Please, choose your weapon: rock, paper or scisors! ')

  if player1.upper() == moves[0].upper() or player1.upper() == moves[1].upper() or player1.upper() == moves[2].upper(): #po "lub" w "if" muszą być ponownie zpisane obydwie strony warunku 

    if player1 == player2:
      print('%s = %s' % (player1, player2))
      print('It is a draw!')

    elif player1 == 'rock':
      if player2 == 'scisors':
        print('%s > %s' % (player1, player2))
        print('Congratulations, you won!')
      else:
        print('%s < %s' % (player1, player2))
        print('Oh no, you lost!')
    
    elif player1 == 'paper':
      if player2 == 'rock':
        print('%s > %s' % (player1, player2))
        print('Congratulations, you won!')
      else:
        print('%s < %s' % (player1, player2))
        print('Oh no, you lost!')

    elif player1 == 'scisors':
      if player2 == 'paper':
        print('%s > %s' % (player1, player2))
        print('Congratulations, you won!')
      else:
        print('%s < %s' % (player1, player2))
        print('Oh no, you lost!')

  elif player1.upper() == 'stop'.upper():
    print('Okay, i get it - you do not like to play with me...')
    break

  else:
    print('Oh, it looks like you picked some move that does not exsist in this game. Try again.')
```
