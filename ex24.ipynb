```
#https://www.practicepython.org/exercise/2014/12/27/24-draw-a-game-board.html

def gameboard(x):
  i = 0
  while i<x:
    print(x*' ---')
    print(x*'|   ' + '|')
    i+=1
  print(x*' ---')

try:
  x = input('How big gameboard would you like to make? ')
  x = int(x)
  gameboard(x)
except ValueError:
  print('Gameboard size should be an intiger.')
```
