```
#http://www.practicepython.org/exercise/2014/03/12/06-string-lists.html

print('Hello. All i can do is to provide you with the feedback whether your word is a palindrome or not.')

def palindrome(word):

  word_list = list(word) #tworzy liste ze znaków danego słowa
  word_list.reverse() #odwraca kolejność znaków w liście

  word_tbc = ''.join(word_list) #łączy znaki listy w słowo, bez znaków pomiędzy elementami listy

  if word_tbc.upper() == word.upper(): #dodanie .upper() do String w tym poleceniu przyrównuje słowa napisane capslockiem, przez co wielkość znaku nie wpływa na bycie palindromem
    print('%s is a palindrome.' % (word.capitalize())) #.capitalize() podnosi pierwszą literę słowa w wyświetlanym output
  else:
    print('%s is not a palindrome.' % (word.capitalize()))

while True:
  word = input('Please, give me a word: ')
  if word.isdigit(): #sprawdza czy dany element jest liczbą
    print('Your input should be string, not number!')
  else: 
    palindrome(word) #wykonuje funkcje opisaną w "def" dla podanego przez użytkownika "word"
    break
```
