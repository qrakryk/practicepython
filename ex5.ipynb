```
#https://www.practicepython.org/exercise/2014/03/05/05-list-overlap.html

import random

#losowe generowanie listy 1
list1 = []
i=0
list_length = random.randint(10,50) #wybiera losową liczbę z zakresu <10,30>
while i<list_length:
  number = random.randint(0,50) #wybiera losową liczbę z zakresu <0,30>
  list1.append(number)
  i+=1

#losowe generowanie listy 2
list2 = []
i=0
list_length = random.randint(10,30) #wybiera losową liczbę z zakresu <10,30>
while i<list_length:
  number = random.randint(0,30) #wybiera losową liczbę z zakresu <0,30>
  list2.append(number)
  i+=1

#sortowanie list "od najmniejszej do największej"
list1.sort()
list2.sort()

#sprawdzenie czy elementy jednej listy zawierają się w drugiej i dodanie wyniku do listy ostatecznej
list_common = []

len1 = len(list1)
len2 = len(list2)

if len1<len2:
  for element in list1:
    if element in list2:
      list_common.append(element)
else:
    for element in list2:
      if element in list1:
        list_common.append(element)

#prezentacja wyniku      
print('First list:')
print(list1)
print('Second list:')
print(list2)
print('Common values:')
print(list_common)
```
