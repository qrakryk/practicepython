```
#https://www.practicepython.org/exercise/2017/02/28/35-birthday-months.html

from collections import Counter #funkcja odpowiedzialna za zliczanie pozycji w liście 

with open('/content/sample_data/bday_list.txt','r') as bday_file:
  content = bday_file.read()

bday_list = content.split('\n')

bday_keys = []
bday_values = []

i = 0
for position in bday_list:
  i=0
  for character in position:
    i+=1
    if character == ':':
      bday_keys.append(position[0:i-1])
      bday_values.append(position[i+1:len(position)])
      break    

bday_dict = dict(zip(bday_keys,bday_values)) #spina dwie równe listy w słownik (klucze,wartości)

print('Welcome to the birthday dictionary. We know the birthday of:')
for key in bday_dict.keys(): 
  print(key)


print('How much people from our list has been born in given months?:')
months=[]
for value in bday_values:
  i = 0
  for character in value:
    i+=1
    if character == '/':
      months.append(value[i:i+2])
      break

c = Counter(months) #zlicza elementy z listy i w formie słownika podaje ile razy dany element występuje na liście, korzysta z biblioteki dodanej na początku ćwiczenia
for element in c:
  print(element,':', c[element])

```
