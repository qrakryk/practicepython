```
#https://www.practicepython.org/exercise/2015/11/01/25-guessing-game-two.html

import random

pc_number = 50
guess = 0 

while True:
  x = input('My number is %d. Is it too high / too low / correct? ' % (pc_number))
  if x == 'high':
    pc_number = pc_number - 1
  if x == 'low':
    pc_number = pc_number + 1
  if x == 'correct':
    guess = guess + 1
    print('Thank you for the game, it took me %d guesses.' % guess)
    break
  guess = guess + 1
```
