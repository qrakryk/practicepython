```
#https://www.practicepython.org/exercise/2014/07/05/18-cows-and-bulls.html

import random

number = random.randint(1000,9999)
attempts = 0

while True:
  try:
    guess = input('Guess my number: ')
    guess = int(guess)
    if guess<1000 or guess>9999:
      print('Your number have to be 4-digit.')
    else:
      number_list = list(str(number))
      guess_list = list(str(guess))
      cows = 0
      bulls = 0
      i=0
      for gue in guess_list:
        if gue == number_list[i]:
          cows+=1
        if (gue != number_list[i]) and (gue in number_list):
          bulls+=1
        i+=1
      attempts+=1
      print('%d cows and %d bulls' % (cows, bulls))
      if cows == 4:
        attempts+=1
        print('You won, my number is %d.' % (number))
        print('It took you %d attempts.' % (attempts))
        break  
  except ValueError:
    print('Your number should be an intiger.')
```
