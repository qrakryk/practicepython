```
#https://www.practicepython.org/exercise/2014/01/29/01-character-input.html

name = input('Name: ')

while True: #pętla działająca aż do polecenia "break", żeby program odpytywał w nieskończoność o wiek aż dostanie intiger
  try: 
    age = input('Age: ')
    age = int(age)
    if age > 0: #wiek musi być liczbą większą od zera
      age_100 = 2020 - age + 100 
      message = '%s, you will reach age of 100 years old in %d year. \n' % (name, age_100) #%s i %d określają pojawianie się zmiennych wspomnianych w "print" po "tekście"; wiadomość może być zapisana w t
      print(message)
      while True:
        try:
          number = input('Okay, now give me another number: ')
          number = int(number)
          print(number*message)
          break
        except ValueError:  
          print('Number must be an intiger. Try again.')
      break #wyjście z pętli True
  except ValueError: #ValueError dotyczy błędu związanego z wartościami w bloku "try" (np. operacja matematyczna nie może być przeprowadzona na string).
    print('Age must be an intiger.') 



```
