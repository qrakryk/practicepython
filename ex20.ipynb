```
#https://www.practicepython.org/exercise/2014/11/11/20-element-search.html

import random

numb_list_lenght = random.randint(10,99)
i = 0
numb_list = []

while i < numb_list_lenght:
  numb = random.randint(1,999)
  if numb not in numb_list:
    numb_list.append(numb)
    i+=1
  
numb_list.sort()

def ismynumber(numbers_list, my_number):
  if my_number in numbers_list:
    message = '%d number is on number list.' % (my_number)
  else:
    message = '%d is not on number list.' % (my_number)
  return print(message)

ismynumber(numb_list, 27)
print(numb_list)
```
