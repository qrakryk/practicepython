```
#http://www.practicepython.org/exercise/2014/05/15/14-list-remove-duplicates.html

a = ['a', 'a', 'b', 'b', 'c', 'c', 'd', 'd', 'e', 'f']

#function #1
def dup_remover1(x):
  b = []
  for element in x:
    if element not in b:
      b.append(element)
  x = b    
  return x

#function2
def dup_remover2(x):
  x = set(x) #set to zbiór elementów bez kolejności. Każdą listę można przerobić na set, a następnie znów na liste, co spowoduje wywalenie duplikatów
  x = list(x)
  return x

print(dup_remover1(a))
print(dup_remover2(a))
```
