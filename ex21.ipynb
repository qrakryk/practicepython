```
#https://www.practicepython.org/solution/2014/07/10/17-decode-a-web-page-solutions.html

import requests
from bs4 import BeautifulSoup
 
base_url = 'https://www.onet.pl/'
r = requests.get(base_url)
soup = BeautifulSoup(r.text)
 
for story_heading in soup.find_all(class_="title"): 
    if story_heading.a: 
        with open('file_to_save.txt', 'a') as open_file: #otwiera plik i w takiej formule kodu nie musze pamiętać żeby go zamknąć, "a" oznacza append i nie nadpisuje danych w pliku
          open_file.write(story_heading.a.text.replace("\n", " ").strip() + '\n') #.write zapisuje w pliku
    else: 
      with open('file_to_save.txt', 'a') as open_file:
        open_file.write(story_heading.contents[0].strip() + '\n')
```
