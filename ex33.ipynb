```
#https://www.practicepython.org/exercise/2017/01/24/33-birthday-dictionaries.html

birth_dict = {
	"Albert Einstein": r'18/11/1881',
	"Nicolai Tesla": r'10/07/1856',
	"Jan Matejko": r'24/06/1838',
	"Benjamin Franklin": r'17/01/1706'
}

print('Welcome to the birthday dictionary. We know the birthday of:')
for key in birth_dict.keys(): #birth_dict.keys() tworzy liste kluczy
  print(key)

check_birth = input('Whose birthday day would you like to know? ')

if check_birth in birth_dict.keys():
  print('%s was born in %s' % (check_birth, birth_dict[check_birth]))
else:
  print('%s is not on our list.' % (check_birth))
```
