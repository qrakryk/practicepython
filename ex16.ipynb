```
#http://www.practicepython.org/exercise/2014/05/28/16-password-generator.html

import random
import os

def password(x):
  if x == 'low':
    words_file = open('/words.txt','r')
    words_content = words_file.read()
    words_list = words_content.split('\n')
    password = random.choice(words_list)
    words_file.close()

  elif x == 'medium':
    words_file = open('/words.txt','r')
    words_content = words_file.read()
    words_list = words_content.split('\n')
    password_word = random.choice(words_list)
    password_number = random.randint(1,999)
    password = '%s%d' % (password_word, password_number)
    words_file.close()

  elif x == 'strong':
    password_lenght = random.randint(9,18)
    i = 0
    password_chars = []
    while i < password_lenght:
      password_chars.append(chr(random.randint(41,126)))
      i+=1
    password = ''.join(password_chars)

  else:
    print('Please, choose one of security standards: low, medium, strong.')

  return print('Your password: %s' % (password))

password('low')
password('medium')
password('strong')



```
